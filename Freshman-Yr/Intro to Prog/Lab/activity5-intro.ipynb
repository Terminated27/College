### A Procedural Programming example

```python
########### Create a new class Employee with attribute name,job,salary
class Employee:
  def __init__(self, my_name, my_job, my_salary): #constructor
    self.name=my_name
    self.job=my_job
    self.salary=my_salary

########## Function
def increase_salary(member,percentage):
  member.salary=member.salary*(1.0+percentage/100.0)

########## Main program
employee1=Employee("Kate","CEO",110000)   # instantiate  employee1

increase_salary(employee1,3.4)            # increase salary by 3.4%
print("New salary of %s, %s, is %s"%(employee1.name,employee1.job,employee1.salary))
```

### From Procedural to Object Oriented Programming

```python
#######################################################
# Create a new class Employee with instance attributes and methods
class Employee:
  def __init__(self, my_name, my_job, my_salary): #constructor
    self.name=my_name
    self.job=my_job
    self.salary=my_salary

  def increase_salary(self,percentage):          # instance method
    self.salary=self.salary*(1.0+percentage/100.0)
#######################################################

##### Main program
employee1=Employee("Kate","CEO",110000)   # instantiate  employee1

employee1.increase_salary(3.4)            # increase salary by 3.4%
print("New salary of %s, %s, is %s"%(employee1.name,employee1.job,employee1.salary))
```

### The magic \_\_str\_\_ method (user's string representation of the object)

```python
#######################################################
# Create a new class Employee with attributes and methods
class Employee:
  def __init__(self, my_name, my_job, my_salary): #constructor
    self.name=my_name
    self.job=my_job
    self.salary=my_salary

  def __str__(self):      # instance method __str__
    return "%s is %s with a salary of %s"%(self.name,self.job,self.salary)

  def increase_salary(self,percentage):      # instance method
    self.salary=self.salary*(1.0+percentage/100.0)
#######################################################

##### Main program
employee1=Employee("Kate","CEO",110000)   # instantiate  employee1

print(employee1)
employee1.increase_salary(3.4)            # increase salary by 3.4%
print("After salary increase: "+str(employee1))
```
