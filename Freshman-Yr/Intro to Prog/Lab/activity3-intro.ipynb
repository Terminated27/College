### Accessing a single or a range of items
#### Extracted subblist notation: A[start:end+1] of size N=(end+1)-start

```python
fruits=["kiwi","mango","banana","cherry","apple"]
print(fruits)
```

```python
print("I like",fruits[2],"but I do not like",fruits[4])
```

```python
print(fruits[0:2],"are very exotic fruits!")
```

```python
print(fruits[2:],"are fruits I eat often!")
```

```python
print(fruits[-1],"is the last fruit in my list!")
```

### changing/adding/removing items

```python
fruits=["kiwi","mango","banana","cherry","apple"]

fruits[2]="pineapple" # replace banana by pineapple
print(fruits)
```

```python
fruits=fruits+["banana"] # add banana at the end
print(fruits)
```

```python
fruits=["papaya"]+fruits # add papaya at the beginning
print(fruits)
```

```python
del(fruits[2:4]) # remove mango and pineapple
print(fruits)
```

### List membership (search)

```python
fruits=["kiwi","mango","banana","cherry","apple"]
print("mango" in fruits)
```

```python
if "mango" in fruits:
  print("I will buy two dozen!")
else:
  print("I will look somewhere else")
```

# For loops

### Using the Range function:
#### Notation:           range(start, end+increment, increment)
####      default values: start=0, increment=1
####      length=(end+increment-start)/increment

### using default range function

```python
sum=0
for i in range(5):
  sum=sum+1
print("the for loop has been executed",sum,'times')
```

```python
for i in range(5):
  print("Index value",i)
```

### using bounds in range function

```python
sum=0
for i in range(2,4):
  print("Index value",i)
  sum=sum+1
print("the for loop has been executed",sum,'times equal to',4-2)
```

### using different increments in range function

```python
sum=0
for i in range(0,10,2):
  print("Index value",i)
  sum=sum+1
print("the for loop has been executed",sum,'times equal to',(10-0)/2)
```

```python
sum=0
for i in range(10,-3,-1):
  print("Index value",i)
  sum=sum+1
print("the for loop has been executed",sum,'times equal to',(-3-10)/(-1))
```

### using a list of items

```python
for item in ["a","e","i","o","u","y"]:
  print(item) 
```

### using a string (scan the letters)

```python
for letter in "aeiouy":
  print(letter)
```

### Scanning a list: indexing vs items 

```python
mylist=[-3.0, 2.0, 3.0, 1.5]

#using direct access to list items
for item in mylist:
  print(item,end=" ")
print()  

#using indexing
for i in range(len(mylist)):
  print(mylist[i],end=" ")
print()

# Remark: strictly speaking it the same thing in Python
# since the range function can be used to return a list of integer using the function "list"
l=list(range(4))
print(l)
```

### Nested for loops

```python
# 3 loops with i,j,k changing
for i in range(4):
  print("i=",i)
  for j in range(3):
    print("\t j=",j)
    for k in range(2):
      print("\t\t k=",k,"\t Hello again")
```

# While loops

### simple use with explicit increment

```python
i=0
while i<5:
  print("Index value",i)
  i=i+1
```

### example of indefinite loop (number of iterations not explicitly known in advance)

```python
num=int(input("Choose a number between 1-100: "))
i=1
while i<num:
  print("Humm, it cannot be",i)
  i=i+1
print("Let me guess, is it",i,"?")
```

# Control statements for loops

### continue statement

```python
# using for loop
mylist=[3,1,9,4,5,7,8,6,2]
for i in range(len(mylist)):
  if (mylist[i]%2!=0):
    continue
  print("I like only even numbers such as",mylist[i])             
```

```python
# this is ok too with for loop
mylist=[3,1,9,4,5,7,8,6,2]
for item in mylist:
  if (item%2!=0):
    continue
  print("I like only even numbers such as",item) 
```

```python
# using while loop
mylist=[3,1,9,4,5,7,8,6,2]
i=0
while i<len(mylist):
  if (mylist[i]%2!=0):
    i=i+1
    continue
  print("I like only even numbers such as",mylist[i])
  i=i+1 
```

### break statement


```python
# using for loop
num=int(input("Choose a number between 1-100: "))

for i in range(100):
  if (i+1)==num:
    break
print("I found it! it is",i+1)
```

```python
#using while loop (loop and a half)
num=int(input("Choose a number between 1-100: "))

i=1
while True:
  if (i==num):
    break
  i=i+1
print("I found it again! it is",i)
```
