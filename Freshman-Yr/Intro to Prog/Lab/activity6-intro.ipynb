# Method objects using String and List
---



### String methods (use return statement)

```python
phrase="Activity 6 is today!"
phrase1=phrase.lower()
print(phrase1)
print(phrase)
```

```python
print(phrase.upper())
print(phrase.isupper())
print(phrase.upper().isupper())
```

```python
print(phrase.strip("!"))
```

```python
phrase="Activity 6 is today!!   "
print(phrase)
phrase=phrase.rstrip().rstrip("!")
print(phrase)
```

```python
print(phrase.replace("is","was"))
```

```python
phrase="Activity 6 is today!"
mylist=phrase.split()
print(mylist)
a,b,c,d=phrase.split()
print(a,b,c,d)
```

### List methods (use return value or work/modify in place)

```python
# append 1st approach (return value)
mylist1=mylist+["!!"] # replace mylist1 by mylist to mimic append
print(mylist1)
# append 2nd approach (work in place- no return/no copy- more efficient)
mylist.append("!!")
print(mylist)
```

```python
mylist.remove("!!") # work in place (search and remove)
print(mylist)
```

```python
mylist.insert(1,"!!") #insert at a particular index
print(mylist)
```

# Files 
### (on colab, files will be created and read from a local folder in the cloud) 

```python
# create a new file
f1=open("data.txt","w")
f1.write("Luke\n")
f1.write("Leia\n")
f1.close() 
```

```python
# append to existing file
f1=open("data.txt","a")
f1.write("Obiwan\n")
f1.write("Yoda\n")
f1.close()
```

```python
# read an existing file - 1st approach
f1=open("data.txt","r")
full_content=f1.read()
##this equivalent to: full_content="Luke\nLeia\n\Obiwan\nYoda\n"
f1.close()
print(full_content)
print(full_content.rstrip()) # strip the last \n
```

```python
# read an existing file - 2nd approach line by line
f1=open("data.txt","r")
for i in range(4):
  line_content=f1.readline().rstrip()
  print(line_content)
f1.close()
```

```python
# read an existing file - 3rd approach line by line (best)
f1=open("data.txt","r")
for line in f1:
  print(line.rstrip())
f1.close()
```

```python
# read an existing file - 4th approach- stores all lines for later use
f1=open("data.txt","r")
lines=f1.readlines() # create a list
f1.close()


for i in range(len(lines)):
  print(lines[i].rstrip())

print()

for line in lines:
  print(line.rstrip())
```
