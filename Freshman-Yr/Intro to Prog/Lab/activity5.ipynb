# Method objects- Application 1

```python
#Q1 [2pts]: Write the class Circle. The output of the program should be:
#          Hint: radius is an attribute (instance variable)
#                compute_perimeter is a method
"""
radius is 3.0
perimeter is 18.84955592153876
"""
import math
class Circle:
  # to complete
  def __init__(self, radius):
    self.radius=radius
  def compute_perimeter(self):
    return 2*math.pi*self.radius
# Main program
circ1=Circle(3.0)
print("radius is",circ1.radius)
print("perimeter is",circ1.compute_perimeter())
```

```python
#Q2 [2pts]: Modify the class Circle (rewrite it below) and include the methods
#          __str__ and scale, to get the following output:
"""
radius is 3.0
radius is 6.0
radius is 3.0
"""
import math
class Circle:
  # to complete
  def __init__(self, radius):
    self.radius=radius
  def __str__(self):
    return f"radius is {self.radius}"
  def scale(self, scaler):
    self.radius = self.radius*scaler
    return self.radius


# Main program
circ1=Circle(3.0)
print(circ1)
circ1.scale(2.0)
print(circ1)
circ1.scale(0.5)
print(circ1)
```

```python
#Q6 [2pts]: Rewrite the class Quad to include the discriminant and 
#solve methods, to obtain: 
#(Requirement: the solve method must call the discriminant method)
"""
(-2)+(3)x+(2)x^2
discriminant is  25
solutions of f(x)=0 are [0.5, -2.0]
Let us check
f(0.5)=0.0
f(-2.0)=0.0
"""
import math

class Quad:
  # to complete
  def __init__(self, c0=0, c1=0, c2=0):
    self.coeff = [c0, c1, c2]
  def __str__(self):
    if self.coeff != [0,0,0]:
      return f"({self.coeff[0]})+({self.coeff[1]})x+({self.coeff[2]})x^2"
    else:
      return "0.0"
  def evaluate(self,x):
    return (self.coeff[0])+(self.coeff[1])*x+(self.coeff[2])*x**2
  def discriminant(self):
    return self.coeff[1]**2-(4*self.coeff[0]*self.coeff[2])
  def solve(self):
      sol = []
      desc = self.discriminant()
      if desc < 0:
        return "No real solutions"
      elif desc == 0:
        sol.append(-self.coeff[1]/(2*self.coeff[0]))
      else:
        pos = -(-self.coeff[1] + math.sqrt(desc))/(2*self.coeff[0])
        neg = -(-self.coeff[1] - math.sqrt(desc))/(2*self.coeff[0])
        sol.append(pos)
        sol.append(neg)
      return sol


  

### Main program
q1=Quad(-2,3,2)
print(q1)
print("discriminant is ",q1.discriminant())
solutions=q1.solve()
print("solutions of f(x)=0 are",solutions)
print("Let us check")
print("f(%s)=%s"%(solutions[0],q1.evaluate(solutions[0])))
print("f(%s)=%s"%(solutions[1],q1.evaluate(solutions[1])))
```
