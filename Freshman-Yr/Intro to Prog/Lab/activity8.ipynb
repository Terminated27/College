###The goal of this section is to implement a class Gaussian that can be used
###to represent a family of Gaussian functions of the form:

```python
#    a*exp(-(x/b)**2)   where a,b are real *non-zero* numbers 
```

### Write the class Gaussian (you need to proceed step by step) -- do not modify the main codes

```python
import math
### class Gaussian here
class Gaussian:
  def __init__ (self, a = 1, b = 2):
    self.a = a
    self.b = b
  def __str__ (self):
    return f"{self.a}*exp(-(x**2/{self.b**2}))"
  def derive(self):
    return f"-{2*self.a/self.b**2}*exp(-(x**2/{self.b**2}))*x"
  def __mul__(self, mul):
    if isinstance(mul, (float,int)):
      return Gaussian(self.a*mul, self.b)
    elif isinstance(mul, Gaussian):
      if self == 0 or mul == 0:
        return "0"
      a = self.a * mul.a
      b = math.sqrt(1/((1/self.b**2) + (1/mul.b**2)))
      return Gaussian(a,b)
  def __truediv__ (self, div):
    if div.a == 0:
      return ("cannot divide by zero")
    elif (1/self.b**2) - (1/div.b**2) == 0:
      return self.a / div.a 
    else:
      a = self.a / div.a 
      b = math.sqrt(1/((1/self.b**2) - (1/div.b**2)))
      return Gaussian(a,b)
  def evaluate(self, x):
      return (self.a)*math.exp(-(x**2/(self.b**2)))
    
  
  ### To complete step-by step

    
```
